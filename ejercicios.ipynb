Usando los métodos estudiados, determine todas las raíces de la ecuación:

```python
from scipy.optimize import newton
import numpy as np

def f(x):
    return x**5 + 5.25*x**4 + 4.125*x**3 - 9.125*x**2 - 14.625*x - 5.625


puntos_suposicion = (-6,6,100)
raices = []


for x0 in puntos_suposicion:
    try:
        r = newton(f, x0)
        r = np.round(r,7)
        if not any(np.isclose(r,raices)):
            raices.append(r)
    except:
        pass

print("Raices:", raices)
for r in raices: 
    print(r)


```
